# RMSD

```python
import pandas as pd
```

```python
rank1 = pd.read_csv("./rank_1_rmsd_results.csv")
rank2 = pd.read_csv("./rank_2_rmsd_results.csv", usecols=["rank_2_ligand"])
rank3 = pd.read_csv("./rank_3_rmsd_results.csv", usecols=["rank_3_ligand"])
rank4 = pd.read_csv("./rank_4_rmsd_results.csv", usecols=["rank_4_ligand"])
rank5 = pd.read_csv("./rank_5_rmsd_results.csv", usecols=["rank_5_ligand"])
rank6 = pd.read_csv("./rank_6_rmsd_results.csv", usecols=["rank_6_ligand"])
rank7 = pd.read_csv("./rank_7_rmsd_results.csv", usecols=["rank_7_ligand"])
rank8 = pd.read_csv("./rank_8_rmsd_results.csv", usecols=["rank_8_ligand"])
rank9 = pd.read_csv("./rank_9_rmsd_results.csv", usecols=["rank_9_ligand"])
rank10 = pd.read_csv("./rank_10_rmsd_results.csv", usecols=["rank_10_ligand"])
```

```python
merge_data = pd.concat([rank1, rank2, rank3, rank4, rank5, rank6, rank7, rank8, rank9, rank10], axis=1)
merge_data.head()
```

```python
merge_data.index.name = "time"
```

```python
merge_data.drop(columns=["time"], inplace=True)
```

```python
merge_data.head()
```

# RMSF

```python
import pandas as pd
```

```python
rank1 = pd.read_csv("./rank1_rmsf_results.csv")
rank2 = pd.read_csv("./rank2_rmsf_results.csv", usecols=["rmsf"])
rank3 = pd.read_csv("./rank3_rmsf_results.csv", usecols=["rmsf"])
rank4 = pd.read_csv("./rank4_rmsf_results.csv", usecols=["rmsf"])
rank5 = pd.read_csv("./rank5_rmsf_results.csv", usecols=["rmsf"])
rank6 = pd.read_csv("./rank6_rmsf_results.csv", usecols=["rmsf"])
rank7 = pd.read_csv("./rank7_rmsf_results.csv", usecols=["rmsf"])
rank8 = pd.read_csv("./rank8_rmsf_results.csv", usecols=["rmsf"])
rank9 = pd.read_csv("./rank9_rmsf_results.csv", usecols=["rmsf"])
rank10 = pd.read_csv("./rank10_rmsf_results.csv", usecols=["rmsf"])
```

```python
merge_data = pd.concat([rank1, rank2, rank3, rank4, rank5, rank6, rank7, rank8, rank9, rank10], axis=1)
merge_data.head()
```

```python
merge_data.index = merge_data["residue"]
```

```python
merge_data.head()
```

```python
merge_data.drop(columns=["residue", "Unnamed: 0"], inplace=True)
merge_data.head()
```

```python
columns_name = ["rank_1_ligand", "rank_2_ligand", "rank_3_ligand", "rank_4_ligand", "rank_5_ligand", "rank_6_ligand", "rank_7_ligand", "rank_8_ligand", "rank_9_ligand", "rank_10_ligand"]
merge_data.columns = columns_name
merge_data.head()
```

```python
import matplotlib.pyplot as plt
```

```python
merge_data.plot(title="RMSF of protein and ligand")
plt.xlabel("Residue number")
plt.ylabel("RMSF (nm)");
plt.legend(bbox_to_anchor=(1.35, 1.02), loc="upper right")
plt.savefig("rmsf_merged_plot.png")
```
